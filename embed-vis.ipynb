```python
 ##### Joint projection of combined embeddings
'''
Fits the dimensionality reduction (UMAP or t-SNE) once on the combined embeddings. 
After fitting, it splits the result back into two parts corresponding to the original inputs.
This approach preserves the relative relationship between the two sets because both are embedded in the same 2D projection space. 
So we can directly compare how the points from emb1 and emb2 relate spatially in the same coordinate system.
Better way to see how the embeddings shift/change relative to each other.
'''

import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import umap
import seaborn as sns

def visualize_embeddings(emb1, labels1, emb2, labels2, label_names=("Before", "After"), method="umap", dataset_name="Dataset", model="Model"):
    if method == "tsne":
        reducer = TSNE(n_components=2, random_state=42)
    elif method == "umap":
        reducer = umap.UMAP(n_components=2, random_state=42)
    else:
        raise ValueError("Choose 'umap' or 'tsne'")

    emb_combined = np.vstack([emb1, emb2])
    proj = reducer.fit_transform(emb_combined)

    n = len(emb1)
    proj1, proj2 = proj[:n], proj[n:]

    plt.figure(figsize=(12, 5))

    for i, (proj, labels, title) in enumerate(zip([proj1, proj2], [labels1, labels2], label_names)):
        plt.subplot(1, 2, i + 1)
        sns.scatterplot(x=proj[:, 0], y=proj[:, 1], hue=labels, palette="Set1", alpha=0.7, s=30)
        plt.title(title)
        plt.xlabel("Component 1")
        plt.ylabel("Component 2")
        plt.legend(title="Class", bbox_to_anchor=(1.05, 1), loc='upper left')
        plt.tight_layout()

    plt.suptitle("{} Embedding Visualization for {} ({})".format(model, dataset_name.upper(), method.upper()), fontsize=16, y=1.05)
    plt.tight_layout()
    plt.show()
```

```python
dataset = "bace"
base_dir = "molclr_results"
baseline_dir = "freeze_"  
best_combo_dir = "freeze_0-4"  

# Load baseline
emb1 = np.load(os.path.join(base_dir, dataset, baseline_dir, "embeddings.npy"))
labels1 = np.load(os.path.join(base_dir, dataset, baseline_dir, "labels.npy")).squeeze()

# Load best-combo
emb2 = np.load(os.path.join(base_dir, dataset, best_combo_dir, "embeddings.npy"))
labels2 = np.load(os.path.join(base_dir, dataset, best_combo_dir, "labels.npy")).squeeze()

# Visualize
visualize_embeddings(emb1, labels1, emb2, labels2,
                     label_names=(baseline_dir, best_combo_dir),
                     method="umap", dataset_name=dataset, model="MolCLR")  

visualize_embeddings(emb1, labels1, emb2, labels2,
                     label_names=(baseline_dir, best_combo_dir),
                     method="tsne", dataset_name=dataset, model="MolCLR")
```

```python
##### Separate projections for each embedding set
'''
Applies dimensionality reduction separately on emb1 and emb2 using independent fits.
The two 2D projections are generated independently and the coordinate systems differ, 
so we cannot directly compare their spatial layouts across the two plots.
The embeddings look different because each is fit separately and the axes are not aligned.
better for when we want to inspect the structure of each embedding separately without comparing them directly.
'''

import matplotlib.pyplot as plt
import seaborn as sns
import umap
from sklearn.manifold import TSNE

def visualize_embeddings2(emb1, labels1, emb2, labels2, label_names=("Baseline", "Best Freeze"), method="umap", dataset_name="Dataset", model="Model"):
    # Reduce dimensionality
    if method == "umap":
        reducer = umap.UMAP(n_neighbors=15, min_dist=0.1, random_state=42)
        proj1 = reducer.fit_transform(emb1)
        proj2 = reducer.fit_transform(emb2)
    elif method == "tsne":
        tsne = TSNE(n_components=2, random_state=42)
        proj1 = tsne.fit_transform(emb1)
        proj2 = tsne.fit_transform(emb2)
    else:
        raise ValueError("Method must be 'umap' or 'tsne'")

    # Ensure labels are 1D for hue
    labels1 = labels1.squeeze()
    labels2 = labels2.squeeze()

    plt.figure(figsize=(14, 6))
    plt.suptitle("{} Embedding Visualization for {} ({})".format(model, dataset_name.upper(), method.upper()), fontsize=16, y=1.05)

    for i, (proj, labels, title) in enumerate(zip([proj1, proj2], [labels1, labels2], label_names)):
        plt.subplot(1, 2, i + 1)
        sns.scatterplot(x=proj[:, 0], y=proj[:, 1], hue=labels, palette="Set1", alpha=0.7, s=30)
        plt.title(title)
        plt.xlabel("Component 1")
        plt.ylabel("Component 2")
        plt.legend(title="Class", loc="best")

    plt.tight_layout(rect=[0, 0, 1, 0.95])  
    plt.show()
```

```python
dataset = "bace"
base_dir = "molclr_results"
baseline_dir = "freeze_"  
best_combo_dir = "freeze_0-4"  

# Load baseline
emb1 = np.load(os.path.join(base_dir, dataset, baseline_dir, "embeddings.npy"))
labels1 = np.load(os.path.join(base_dir, dataset, baseline_dir, "labels.npy")).squeeze()

# Load best-combo
emb2 = np.load(os.path.join(base_dir, dataset, best_combo_dir, "embeddings.npy"))
labels2 = np.load(os.path.join(base_dir, dataset, best_combo_dir, "labels.npy")).squeeze()

# Visualize
visualize_embeddings2(emb1, labels1, emb2, labels2,
                     label_names=(baseline_dir, best_combo_dir),
                     method="umap", dataset_name=dataset, model="MolCLR")  

visualize_embeddings2(emb1, labels1, emb2, labels2,
                     label_names=(baseline_dir, best_combo_dir),
                     method="tsne", dataset_name=dataset, model="MolCLR")
```
